### 2.1 Recursive Factorial
Write a recursive function to calculate the factorial of a given number.

```python
# Recursive function to calculate factorial

def factorial(n):
    if n == 0 or n == 1:   # Base case
        return 1
    else:
        return n * factorial(n - 1)   # Recursive case

# input number
num = int(input("Enter a number: "))

print("The factorial of", num, "is:", factorial(num))
```

### 2.2 Palindrome Linked List

Write a program to determine if a given linked list is a palindrome.

```python
# Function to check if a linked list (represented as a Python list) is a palindrome
def is_palindrome_linked_list(values):
    return values == values[::-1]

# Function to display the linked list in arrow format
def display_linked_list(values):
    print(" -> ".join(map(str, values)))
```

### 2.3   Merge Sorted Arrays

Write a function that takes two sorted arrays and merges them into a single sorted array.

```python
def merge_sorted_arrays(arr1, arr2):
    i, j = 0, 0
    merged = []
    
    # Traverse both arrays
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1
    
    merged.extend(arr1[i:])
    merged.extend(arr2[j:])
    
    return merged

# array
arr1 = [1, 3, 5]
arr2 = [2, 4, 6]
print(merge_sorted_arrays(arr1, arr2)) 
```

### 2.4   Binary Search Tree
Implement a Binary Search Tree (BST) data structure, including methods for insertion, deletion, and search.

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


class BST:
    def __init__(self):
        self.root = None

    # Insert a new key
    def insert(self, root, key):
        if root is None:
            return Node(key)
        if key < root.key:
            root.left = self.insert(root.left, key)
        elif key > root.key:
            root.right = self.insert(root.right, key)
        return root

    # Search a key
    def search(self, root, key):
        if root is None or root.key == key:
            return root
        if key < root.key:
            return self.search(root.left, key)
        return self.search(root.right, key)

    # Find minimum value node
    def min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    # Delete a key
    def delete(self, root, key):
        if root is None:
            return root

        if key < root.key:
            root.left = self.delete(root.left, key)
        elif key > root.key:
            root.right = self.delete(root.right, key)
        else:
            # Node with one child or no child
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left

            # Node with two children
            temp = self.min_value_node(root.right)
            root.key = temp.key
            root.right = self.delete(root.right, temp.key)

        return root

    # Inorder traversal (sorted order)
    def inorder(self, root):
        if root:
            self.inorder(root.left)
            print(root.key, end=" ")
            self.inorder(root.right)


# usage
bst = BST()
root = None
values = [50, 30, 70, 20, 40, 60, 80]

# Insert values
for v in values:
    root = bst.insert(root, v)

print("Inorder traversal of BST:")
bst.inorder(root) 

# Search for a value
print("\n\nSearching for 40:", "Found" if bst.search(root, 40) else "Not Found")

# Delete a value
root = bst.delete(root, 20)
print("\nInorder after deleting 20:")
bst.inorder(root) 
```

### 2.5   Longest Palindromic Substring

Write a program to find the longest palindromic substring within a given string.


```python
def longest_palindromic_substring(s):
    if not s:
        return ""

    start, end = 0, 0  # Track best substring

    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return left + 1, right - 1  # return valid bounds

    for i in range(len(s)):
        # Odd length palindrome
        l1, r1 = expand_around_center(i, i)
        # Even length palindrome
        l2, r2 = expand_around_center(i, i + 1)

        # Pick the longer palindrome
        if r1 - l1 > end - start:
            start, end = l1, r1
        if r2 - l2 > end - start:
            start, end = l2, r2

    return s[start:end + 1]


# usage
print(longest_palindromic_substring("babad"))  
print(longest_palindromic_substring("cbbd")) 
```

### 2.6   Merge Intervals

Write a program to merge overlapping intervals in a list of intervals.

```python
def merge_intervals(intervals):
    if not intervals:
        return []

    # Sort by start time
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]

    for current in intervals[1:]:
        prev = merged[-1]

        # Check overlap
        if current[0] <= prev[1]:
            # Merge by extending the end
            merged[-1] = (prev[0], max(prev[1], current[1]))
        else:
            merged.append(current)

    return merged


# usage
intervals = [(1, 3), (2, 6), (8, 10), (15, 18)]
print(merge_intervals(intervals))  
```

### 2.7  Maximum Subarray

Write a program to find the maximum sum of a contiguous subarray within a given array.


```python
def max_subarray(nums):
    # Initialize with the first element
    max_current = max_global = nums[0]

    for num in nums[1:]:
        # Either extend the subarray or start a new one
        max_current = max(num, max_current + num)
        # Update global maximum
        max_global = max(max_global, max_current)

    return max_global


# usage
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_subarray(arr))
```

### 2.8  Reverse Linked List

Write a program to reverse a singly-linked list.

```python
# Node definition
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class LinkedList:
    def __init__(self):
        self.head = None

    # Insert new node at the end
    def append(self, val):
        new_node = ListNode(val)
        if not self.head:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

    # Print linked list
    def print_list(self):
        curr = self.head
        while curr:
            print(curr.val, end=" -> " if curr.next else "")
            curr = curr.next
        print()

    # Reverse linked list (iterative)
    def reverse(self):
        prev, curr = None, self.head
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        self.head = prev


# usage
ll = LinkedList()
for val in [1, 2, 3, 4, 5]:
    ll.append(val)

print("Original list:")
ll.print_list()   

ll.reverse()
print("Reversed list:")
ll.print_list()  
```

### 2.9 Minimum Edit Distance

Write a program to calculate the minimum number of operations (insertions, deletions, or substitutions) required to transform one string into another.

```python
def min_edit_distance(word1, word2):
    m, n = len(word1), len(word2)

    # DP table of size (m+1) x (n+1)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # cases
    for i in range(m + 1):
        dp[i][0] = i  # Deleting all characters
    for j in range(n + 1):
        dp[0][j] = j  # Inserting all characters
        
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] 
            else:
                dp[i][j] = 1 + min(
                    dp[i - 1][j],    # Deletion
                    dp[i][j - 1],    # Insertion
                    dp[i - 1][j - 1] # Substitution
                )

    return dp[m][n]


# usage
print(min_edit_distance("kitten", "sitting")) 
```

### 2.10  Boggle Game

Implement a program that solves the Boggle game, given a board and a list of words.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            node = node.children[ch]
        node.is_end_of_word = True


class BoggleSolver:
    def __init__(self, board, words):
        self.board = board
        self.rows = len(board)
        self.cols = len(board[0])
        self.trie = Trie()
        for word in words:
            self.trie.insert(word)
        self.found_words = {}

    def find_words(self):
        visited = [[False for _ in range(self.cols)] for _ in range(self.rows)]
        for r in range(self.rows):
            for c in range(self.cols):
                self._dfs(r, c, self.trie.root, "", visited, [])
        return self.found_words

    def _dfs(self, r, c, node, prefix, visited, path):
        if r < 0 or c < 0 or r >= self.rows or c >= self.cols or visited[r][c]:
            return

        char = self.board[r][c]
        if char not in node.children:
            return

        visited[r][c] = True
        node = node.children[char]
        new_prefix = prefix + char
        path.append((r, c))

        if node.is_end_of_word and new_prefix not in self.found_words:
            self.found_words[new_prefix] = path.copy()

        # Explore all 8 directions
        for dr in [-1, 0, 1]:
            for dc in [-1, 0, 1]:
                if dr != 0 or dc != 0:
                    self._dfs(r + dr, c + dc, node, new_prefix, visited, path)

        visited[r][c] = False
        path.pop()


# usage
board = [
    ['o', 'a', 'a', 'n'],
    ['e', 't', 'a', 'e'],
    ['i', 'h', 'k', 'r'],
    ['i', 'f', 'l', 'v']
]
words = ["oath", "pea", "eat", "rain", "hike"]

solver = BoggleSolver(board, words)
results = solver.find_words()

print("Words found:")
for word, path in results.items():
    print(f"{word}")
```

### 2.11  Fibonacci Generator 

Write a generator function that produces Fibonacci numbers using yield.

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


# usage
for num in fibonacci(10):
    print(num)
```

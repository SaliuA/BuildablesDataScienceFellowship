## 2. Statistical Analysis and Probability Distributions with the Iris Dataset 

### 2.1 Descriptive Statistics:

Descriptive Statistics is a branch of statistics that deals with summarizing and presenting data in a meaningful way. It helps to organize raw data into simple forms that are easy to understand without making predictions or drawing conclusions beyond the data.

### 2.1.1 Central Tendencies
Calculating the mean, median, and mode of the sepal
lengths.

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))

# Plot KDE of sepal length
sns.kdeplot(df['sepal length (cm)'], color='orange', linewidth=2, alpha=0.7)

# Compute mean, median, mode
mean_val = df['sepal length (cm)'].mean()
median_val = df['sepal length (cm)'].median()
mode_val = df['sepal length (cm)'].mode()[0]

# Plot mean
plt.axvline(mean_val, color='blue', linestyle='--', linewidth=2, label=f"Mean: {mean_val:.2f}")
plt.text(mean_val + 0.1, plt.ylim()[1]*0.9, f"Mean: {mean_val:.2f}", color='blue')

# Plot median
plt.axvline(median_val, color='purple', linestyle='--', linewidth=2, label=f"Median: {median_val:.2f}")
plt.text(median_val + 0.1, plt.ylim()[1]*0.85, f"Median: {median_val:.2f}", color='purple')


# Formatting
plt.title("KDE of Sepal Length in Iris Dataset", fontsize=14, fontweight="bold")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.show()

print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Mode: {mode_val:.2f} of the sepal length")

```

fig 1.0 above shows the closeness between the mean and median values of 5.84, 5.8 respectively. This indicate a dataset without much variations in the sepal length

### 2.2 Probability & Distributions:

Probability is the measure of how likely an event is to occur, expressed between 0 (impossible) and 1 (certain). It helps us quantify uncertainty in data and real-world events.

A probability distribution describes how values in a dataset are spread and the likelihood of each outcome. It provides a model of the data’s behavior.

### 2.2.1   Probability Distribution of Sepal Length
Define a random variable for sepal length and
calculate its probability distribution.

```python
# Take sepal length
sepal_lengths = df['sepal length (cm)']

# Create frequency distribution (binning values)
freq_table = sepal_lengths.value_counts(normalize=True).sort_index()

# Convert to probability distribution
prob_dist = freq_table.reset_index()
prob_dist.columns = ['Sepal Length (cm)', 'Probability']
prob_dist.head(10)
```

### 2.2.3 Cumulative Distribution Function (CDF)
Calculating and plotting the CDF for petal lengths.

```python
plt.figure(figsize=(12,6))

# Plot CDF using Seaborn
sns.ecdfplot(data=df, x="petal length (cm)", color="darkorange", linewidth=2)

# Labels and title
plt.title("Cumulative Distribution Function (CDF) of Petal Lengths", fontsize=14, fontweight="bold")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Cumulative Probability")
plt.grid(True, linestyle="--", alpha=0.6)
#show the plow
plt.tight_layout()
plt.show()
```

fig 3.0 The Cumulative Distribution Function (CDF) shows the probability that a random variable takes a value less than or equal to a given point. From the curve, we can see that shorter petal lengths are less common, while most values lie between 3–5 cm, where the curve rises steeply. The CDF provides an easy way to understand the distribution and proportion of data below any chosen value.

### 2.2.5 Probability Estimation
Determine the probability that a randomly selected iris
flower has a petal length greater than a given threshold (e.g., 4.5
cm).

```python
# Threshold
threshold = 4.5

# Calculate probability
total = len(df)
greater_count = (df['petal length (cm)'] > threshold).sum()
probability = greater_count / total

print(f"Probability that petal length > {threshold} cm: {probability:.3f}")
```

This indicates there is a 42 % chance of selecting an iris flower with a petal length greater than 4.5cm

### 2.3 Inferential Statistics:

Inferential statistics is a branch of statistics that allows us to make conclusions, predictions, or generalizations about a larger population based on data collected from a sample. Unlike descriptive statistics, which only summarizes data (e.g., mean, median, variance), inferential statistics uses probability theory and statistical models to test hypotheses, estimate population parameters, and determine relationships between variables.

### 2.3.1 Hypothesis Testing
Perform a t-test (or ANOVA) to check if there is
a significant difference in mean petal length between two iris
species (e.g., Setosa vs Versicolor).

```python

from scipy.stats import ttest_ind

# Add species names to the dataframe
df['species'] = df['target'].apply(lambda x: iris.target_names[x])

# Extract petal length for Setosa and Versicolor
setosa = df[df['species'] == 'setosa']['petal length (cm)']
versicolor = df[df['species'] == 'versicolor']['petal length (cm)']

# Perform independent t-test
t_stat, p_value = ttest_ind(setosa, versicolor)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4e}")
```

The t-test produced a t-statistic of -39.49 and a p-value of 5.40 × 10⁻⁶², which is extremely small (almost zero). This result strongly indicates that the difference in mean petal length between Setosa and Versicolor is statistically significant. In other words, the probability that this difference occurred by random chance is practically zero, confirming that the two species have distinct petal lengths.

### 2.3.2 Covariance & Correlation
Calculate and interpret the covariance and correlation
coefficient between sepal length and sepal width.

```python
# Calculate covariance
covariance = df['sepal length (cm)'].cov(df['sepal width (cm)'])

# Calculate correlation
correlation = df['sepal length (cm)'].corr(df['sepal width (cm)'])

print("Covariance:", covariance)
print("Correlation:", correlation)
```

1. Covariance: -0.0424

Covariance measures the direction of the relationship between two variables (but not the strength).

A small negative covariance means that as sepal length increases, sepal width tends to decrease slightly.

However, since the value is very close to 0, the relationship is weak.

2. Correlation: -0.118

Correlation is a standardized measure (ranges from -1 to +1).

A correlation of -0.118 suggests a very weak negative linear relationship between sepal length and sepal width.

This means that knowing the sepal length doesn’t strongly help in predicting sepal width, and vice versa.
